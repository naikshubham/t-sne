# t-sne on Credit card fraud detection Dataset

### Color Coding :
#### Orange : Class 1
#### Blue   : Class 0

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('creditcard.csv')
real_trans = data[data['Class'] == 0]                   #extract real transcations
fraud_trans = data[data['Class'] == 1]                  #extract fraud transcations
data.shape
```

```python
#sample the dataset to get equal proportion of real and fraud transcations
def sample(n):
    fraud_trans_sample = round(n * 0.00172)                         #as % of fraud trans in dataset is 0.00172
    real_trans_sample = n - fraud_trans_sample                      #real_trans sample is total - fraud_trans
    sample_fraud_trans = fraud_trans.sample(n=fraud_trans_sample)   #create sampled DataFrame for fraud trans
    sample_real_trans = real_trans.sample(n=real_trans_sample)      #create sampled DataFrame for real trans
    sampled_data = sample_fraud_trans.append(sample_real_trans)     #create final DataFrame with fraud aswell as real trans
    return sampled_data
```

For perplexity = 30 and step =5000 on 10000 samples

```python
data = sample(10000)
label = data['Class']                                     #extract class labels in Label
standardized_data = StandardScaler().fit_transform(data)  #normalize the data
std_data = standardized_data
```

```python
#t-sne using scikit learn
model = TSNE(n_components=2,perplexity=30,random_state=0,n_iter=5000) 
# configuring the parameteres
# the number of components = 2
# default perplexity = 30
# default learning rate = 200
# default Maximum number of iterations for the optimization = 5000

tsne_data = model.fit_transform(std_data)

#creating a new data frame which will help in plotting results
tsne_data = np.vstack((tsne_data.T,label)).T
tsne_df = pd.DataFrame(data=tsne_data,columns = ('Dim_1','Dim_2','label'))

#plotting the results of tsne
sns.FacetGrid(tsne_df,hue = 'label',size =6).map(plt.scatter,'Dim_1','Dim_2').add_legend()
plt.show()
```
